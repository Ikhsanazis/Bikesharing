```python
df["dteday"]=pd.to_datetime(df["dteday"])
```

```python
df.info()
```

```python
#rename the column name
new_col_name={
    "dteday":"date",
    "yr":"year",
    "mnth":"month",
    "hum":"humidity",
    "cnt":"count"
}

df.rename(new_col_name,axis=1,inplace=True)
```

```python
#change the data in column season
new_season={"1":"spring",
            "2":"summer",
            "3":"fall",
            "4":"winter"}


df.season=df.season.astype("str")
df.season=df.season.map(new_season)
```

```python
#change the data in column weathersit
new_weathersit={"1":"Clear",
                "2":"Mist",
                "3":"Light Snow",
                "4":"Heavy Rain"}

df.weathersit=df.weathersit.astype("str")
df.weathersit=df.weathersit.map(new_weathersit)
print(new_weathersit)
```
